```python
# Global Solution - Dynamic Programming
# Integrantes:
# - Caio Berardo de Araújo, RM: 560357
# - Giovanni Romano Provazi, RM: 560434
# - Guilherme Augusto Caseiro, RM: 559765
# - Leonardo Fernandes Mesquita, RM: 559623
# - Vitor de Lima Domingues, RM: 561008

# Bibliotecas
import math
import time


# Sistema do Satelite
class Satelite:

    def __init__(self, name, x, y, battery):
        self.name = name
        self.x = x  # Coordenada X simulada no espaço
        self.y = y  # Coordenada Y simulada no espaço
        self.battery = battery  # Nível de bateria (0 a 100%)

    def status(self):
        return f"{self.name:15} | Posição: ({self.x:3}, {self.y:3}) | Bateria: {self.battery}%"


# Sistema de Logistica
class SpaceLogisticalSystem:

    def __init__(self):
        self.constelation = []

    def add_satelite(self, satelite):
        self.constelation.append(satelite)
        print(f"\n[SUCESSO] {satelite.name} integrado à rede de monitoramento.")

    def list_satelites(self):
        if not self.constelation:
            print("\n[AVISO] Nenhum satélite cadastrado na rede ainda.")
            return
        print("\n--- SATÉLITES EM ÓRBITA ---")
        for sat in self.constelation:
            print(sat.status())
        print("-" * 50)

    def calculate_distance(self, sat_x, sat_y, est_x, est_y):
        # Usando a fórmula da distância euclidiana/pitagórica para simular proximidade
        return math.sqrt((sat_x - est_x) ** 2 + (sat_y - est_y) ** 2)

    def optimize_conection(self, station_name, station_x, station_y):
        if not self.constelation:
            print(
                "\n[ERRO] Impossível otimizar. Nenhum satélite cadastrado no sistema."
            )
            return

        print(
            f"\n[SISTEMA] Buscando melhor satélite para conexão com: {station_name}..."
        )
        time.sleep(0.8)  # Efeito dramático para simular processamento

        best_satelite = None
        lowest_distance = float("inf")

        for sat in self.constelation:
            # Regra de negócio 1: Satélites com menos de 20% de bateria não são confiáveis
            if sat.battery < 20:
                print(
                    f"   * {sat.name} descartado: Bateria baixa ({sat.battery}%)."
                )
                continue

            dist = self.calculate_distance(sat.x, sat.y, station_x, station_y)
            print(f"   * {sat.name} está a uma distância de {dist:.2f} unidades.")

            # Regra de negócio 2: Escolher o satélite mais próximo com bateria suficiente
            if dist < lowest_distance:
                lowest_distance = dist
                best_satelite = sat

        print("\n" + "=" * 50)
        if best_satelite:
            print(f"   SOLUÇÃO OTIMIZADA ENCONTRADA!")
            print(f"   Satélite Recomendado: {best_satelite.name}")
            print(f"   Distância: {lowest_distance:.2f} unidades")
            print(f"   Status da Bateria: {best_satelite.battery}%")
        else:
            print(
                "   [ERRO] Nenhum satélite disponível ou em condições operacionais na região!"
            )
        print("=" * 50)


def show_menu():
    print("\n" + "#" * 40)
    print("      SPACE CONNECT - GLOBAL SOLUTION")
    print("#" * 40)
    print("[1] Cadastrar Novo Satélite")
    print("[2] Listar Constelação Atual")
    print("[3] Otimizar Conexão com Estação Terrestre")
    print("[0] Sair do Sistema")
    print("#" * 40)


# Loop principal do sistema interativo
if __name__ == "__main__":
    # Inicializando o sistema de logística
    system = SpaceLogisticalSystem()

    # Dados iniciais pré-carregados para demonstração
    system.constelation.append(Satelite("FIAP-Sat Alpha", x=12, y=35, battery=85))
    system.constelation.append(Satelite("FIAP-Sat Beta", x=45, y=50, battery=15))
    system.constelation.append(Satelite("FIAP-Sat Gamma", x=18, y=22, battery=90))

    while True:
        show_menu()
        option = input("Escolha uma opção (0-3): ").strip()

        # [1] Cadastrar Novo Satélite
        if option == "1":
            print("\n--- CADASTRO DE NOVO SATÉLITE ---")
            try:
                name = input("Nome do Satélite: ").strip()
                x = int(input("Coordenada X da órbita: "))
                y = int(input("Coordenada Y da órbita: "))
                battery = int(input("Porcentagem da Bateria (0-100): "))

                if 0 <= battery <= 100:
                    new_sat = Satelite(name, x, y, battery)
                    system.add_satelite(new_sat)
                else:
                    print("\n[ERRO] A bateria deve ser um valor entre 0 e 100.")
            except ValueError:
                print(
                    "\n[ERRO] Entrada inválida! Coordenadas e bateria devem ser números inteiros."
                )

        # [2] Listar Constelação Atual
        elif option == "2":
            system.list_satelites()

        # [3] Otimizar Conexão com Estação Terrestre
        elif option == "3":
            print("\n--- LOCALIZAÇÃO DA ESTAÇÃO TERRESTRE ---")
            try:
                name_station = input("Nome da Estação (ex: Base Amazônia): ").strip()
                station_x = int(input("Digite a coordenada X da estação: "))
                station_y = int(input("Digite a coordenada Y da estação: "))

                system.optimize_conection(name_station, station_x, station_y)
            except ValueError:
                print(
                    "\n[ERRO] Entrada inválida! As coordenadas devem ser números inteiros."
                )

        # [0] Sair do Sistema
        elif option == "0":
            print("\nDesconectando do Space Connect... Sistema encerrado.")
            break

        # Opção Inválida
        else:
            print(
                "\n[AVISO] Opção inválida! Por favor, escolha um número de 0 a 3."
            )

        # Pausa necessária para leitura no terminal antes de limpar/reiniciar o menu
        input("\nPressione ENTER para continuar...")
```
